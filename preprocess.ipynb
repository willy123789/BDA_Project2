```python
# xlsx to csv
def xlsx_to_csv(path):
    wb = openpyxl.load_workbook(path)
    sheet = wb.active
    csv_path = path.replace('.xlsx', '.csv')
    with open(csv_path, 'w') as f:
        c = csv.writer(f)
        for r in sheet.rows:
            c.writerow([cell.value for cell in r])
```
